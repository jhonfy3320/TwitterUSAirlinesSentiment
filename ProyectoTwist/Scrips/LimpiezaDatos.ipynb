```python
from IPython.display import display
# Resumen de Filas y Columnas
shape_info = pd.DataFrame({
    'Filas': [df.shape[0]],
    'Columnas': [df.shape[1]]
})
display(shape_info)

# Tipos de Datos por Columna
dtypes_info = pd.DataFrame(df.dtypes, columns=['Tipo']).reset_index().rename(columns={'index': 'Columna'})
display(dtypes_info)

# Valores Faltantes por Columna
missing_info = df.isna().sum().sort_values(ascending=False).reset_index()
missing_info.columns = ['Columna', 'Nulos']
missing_info = missing_info[missing_info['Nulos'] > 0]
display(missing_info)
```

```python
# Filtrar tweets con sentimiento negativo
df_negativos = df[df['airline_sentiment'] == 'negative']

# Eliminar los registros que no tienen razón de queja
df_negativos = df_negativos[df_negativos['negativereason'].notna()]

# Eliminar columnas irrelevantes o con muchos nulos
columnas_a_eliminar = [
    'tweet_coord',
    'airline_sentiment_gold',
    'negativereason_gold',
    'name'
]
df_negativos = df_negativos.drop(columns=columnas_a_eliminar)

# Reiniciar el índice por orden
df_negativos = df_negativos.reset_index(drop=True)

# Mostrar resumen post-limpieza
print("Forma final del DataFrame limpio:", df_negativos.shape)
print("Columnas conservadas:", df_negativos.columns.tolist())

# Vista previa
df_negativos.head()
```
